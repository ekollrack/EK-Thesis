```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('game_attention.csv')
```

```python
# Home team win: 1 if score_diff > 0
df["home_win"] = (df["score_differential"] > 0).astype(int)

# Away team win: 1 if score_diff < 0
df["away_win"] = (df["score_differential"] < 0).astype(int)

home_games = df[["season", "home_team", "home_win"]].rename(
    columns={"home_team": "team", "home_win": "win"}
)

away_games = df[["season", "away_team", "away_win"]].rename(
    columns={"away_team": "team", "away_win": "win"}
)

team_games = pd.concat([home_games, away_games], ignore_index=True)

team_season_win_pct = (
    team_games
    .groupby(["season", "team"])
    .agg(
        games=("win", "size"),
        wins=("win", "sum")
    )
    .reset_index()
)

team_season_win_pct["win_pct"] = team_season_win_pct["wins"] / team_season_win_pct["games"]

print(team_season_win_pct)

```

```python
# save season, team, games, wins, win_pct to csv
team_season_win_pct.to_csv('team_season_win_pct.csv', index=False)
```
