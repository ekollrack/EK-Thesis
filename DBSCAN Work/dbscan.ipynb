```python
import pandas as pd
from datetime import timedelta
from data_mountain_query.query import get_ambient_tweets
from data_mountain_query.connection import get_connection
from geopy.geocoders import Nominatim
import plotly.express as px
import time
from sklearn.cluster import DBSCAN
import numpy as np
```

```python
games = pd.read_csv("/Users/elisabethkollrack/Thesis/EK-thesis/games.csv")
games = games[games['game_type'] == 'REG']
games['gameday'] = pd.to_datetime(games['gameday'], format='%m/%d/%y')

# Dallas 2013 games
dallas_games = games[
    (games['season'] == 2013) &
    ((games['home_team'] == 'DAL') | (games['away_team'] == 'DAL'))
].sort_values(['season', 'gameday'])
```

```python
collection, client = get_connection(geotweets=True)
```

```python
all_tweets = []
for index, game in dallas_games.iterrows():
    gameday = game['gameday']
    
    # Determine opponent
    if game['home_team'] == 'DAL':
        opponent = game['away_team']
    else:
        opponent = game['home_team']

    anchors = [
        "#DallasCowboys", 
        "#CowboysNation",
        f"#DALvs{opponent}",
        f"#{opponent}vsDAL"
    ]

    start_date = gameday - timedelta(days=3)
    end_date = gameday + timedelta(days=3)
    dates = pd.date_range(start_date, end_date, freq='D')

    for anchor in anchors:
        tweets_list1 = [t for t in get_ambient_tweets(anchor, dates, collection)]
        all_tweets.extend(tweets_list1)

```

```python
geo_df = pd.DataFrame(all_tweets)
print(f"\nTotal geotagged tweets collected: {len(geo_df)}")
geo_df.head(1)

# Extract lon and lat from the 'geo' column
geo_df['lon'] = geo_df['geo'].apply(lambda x: x['coordinates'][0] if isinstance(x, dict) else None)
geo_df['lat'] = geo_df['geo'].apply(lambda x: x['coordinates'][1] if isinstance(x, dict) else None)
```

```python
geo_df.head()
```

```python
coords = geo_df[['lat', 'lon']].to_numpy()

eps_km = 75
eps = eps_km * 1000 / 6371000  # Convert to radians
db = DBSCAN(eps=eps, min_samples=25, algorithm='ball_tree', metric='haversine').fit(np.radians(coords))
cluster_labels = db.labels_
geo_df['cluster'] = cluster_labels
num_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
print(f'Number of clusters with {eps_km} km epsilon: {num_clusters}')
```

```python
# print number of tweets in each cluster
cluster_counts = geo_df['cluster'].value_counts().sort_index()
print(cluster_counts)
```

```python
# parameter sweep for dbscan
for eps_km in [25, 50, 75, 100]:
    eps = eps_km * 1000 / 6371000  # Convert to radians
    db = DBSCAN(eps=eps, min_samples=15, algorithm='ball_tree', metric='haversine').fit(np.radians(coords))
    cluster_labels = db.labels_
    geo_df['cluster'] = cluster_labels
    num_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
    print(f'Number of clusters with {eps_km} km epsilon: {num_clusters}')
    print('Number of tweets in noise cluster:', sum(cluster_labels == -1))
    print('Percentage of tweets in noise cluster:', sum(cluster_labels == -1) / len(cluster_labels) * 100)
    

```

```python
fig = px.scatter_geo(
    geo_df,
    lat='lat',
    lon='lon',
    color='cluster',
    title='DBSCAN Clustering of Geotagged Tweets',
    projection="albers usa",
    scope='usa'
)

fig.update_geos(
    showcountries=True,
    countrycolor="black",
    showsubunits=True,           # Show states
    subunitcolor="gray"          # State boundary color
)

fig.show()
```
