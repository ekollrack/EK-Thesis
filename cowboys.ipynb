```python
import pandas as pd
from datetime import timedelta
from data_mountain_query.query import get_ambient_tweets
from data_mountain_query.connection import get_connection
from geopy.geocoders import Nominatim
import plotly.express as px
import time
```

```python
games = pd.read_csv("/Users/elisabethkollrack/Thesis/EK-thesis/games.csv")
games = games[games['game_type'] == 'REG']
games['gameday'] = pd.to_datetime(games['gameday'], format='%m/%d/%y')

# Dallas 2013 games
dallas_games = games[
    (games['season'] == 2013) &
    ((games['home_team'] == 'DAL') | (games['away_team'] == 'DAL'))
].sort_values(['season', 'gameday'])
```

```python
collection, client = get_connection(user_loc=True)
```

```python
all_tweets = []
for index, game in dallas_games.iterrows():
    gameday = game['gameday']
    
    # Determine opponent
    if game['home_team'] == 'DAL':
        opponent = game['away_team']
    else:
        opponent = game['home_team']

    anchors = [
        "#DallasCowboys", 
        "#CowboysNation",
        f"#DALvs{opponent}",
        f"#{opponent}vsDAL"
    ]

    start_date = gameday - timedelta(days=3)
    end_date = gameday + timedelta(days=3)
    dates = pd.date_range(start_date, end_date, freq='D')

    for anchor in anchors:
        tweets_list = [t for t in get_ambient_tweets(anchor, dates, collection)]
        all_tweets.extend(tweets_list)

```

```python
# Analyze combined city data for all games
tweets_df = pd.DataFrame(all_tweets)

if "city_state" in tweets_df.columns:
    city_counts = (
        tweets_df["city_state"]
        .dropna()
        .astype(str)
        .str.strip()
        .value_counts()
        .reset_index()
    )

# combine NYC and NY
nyc_count = city_counts[city_counts['city_state'] == 'New York, NY']['count'].values[0]
nyc_count += city_counts[city_counts['city_state'] == 'New York City, NY']['count'].values[0]
city_counts = city_counts[city_counts['city_state'] != 'New York City, NY']
city_counts.loc[city_counts['city_state'] == 'New York, NY', 'count'] = nyc_count
city_counts[city_counts['city_state'] == 'New York, NY']

# sort cities
city_counts = city_counts.sort_values(by='count', ascending=False).reset_index(drop=True)

print(f"\nTotal tweets collected: {len(all_tweets)}")
print(f"Total unique cities: {len(city_counts)}")
print(f"Sum of all city counts (should equal len(all_tweets)): {city_counts['count'].sum()}")
print("\nTop 20 cities:")
print(city_counts.head(20))

```

```python
# Initialize geocoder
geolocator = Nominatim(user_agent="tweet_analysis")

def get_coordinates(city_state):
    try:
        # Add USA to improve geocoding accuracy
        location = geolocator.geocode(city_state + ", USA", timeout=10)
        if location:
            return location.latitude, location.longitude
        else:
            print(f"Could not geocode: {city_state}") # safety
            return None, None
    except Exception as e:
        print(f"Error geocoding {city_state}: {e}") # safety
        return None, None

# Copy city_counts
cities = city_counts.copy()

# Get only the cities with at least 100 tweets
cities = cities[cities['count'] >= 50].reset_index(drop=True)

# Get coordinates for each city with a delay 
coordinates = []
for city in cities['city_state']:
    lat, lon = get_coordinates(city)
    coordinates.append((lat, lon))
    time.sleep(1) 

# Split coordinates into separate columns
cities[['lat', 'lon']] = pd.DataFrame(coordinates, index=cities.index)

# Remove any cities that couldn't be geocoded
cities = cities.dropna(subset=['lat', 'lon'])
print(f"Geocoded {len(cities)} cities")

fig = px.scatter_geo(cities,
                     lat="lat",
                     lon="lon",
                     size="count",
                     hover_name="city_state",
                     hover_data={"count": True, "lat": False, "lon": False},
                     title="Dallas Cowboys Tweet Activity by City (2013)",
                     scope="usa")

fig.update_geos(
    scope="usa"
)

fig.update_layout(
    title_x=0.5,
    title_font_size=20
)

fig.show()
```

```python
# save the plot
fig.write_html("dallas_cowboys_tweet_activity_2013.html")
# save coordinates for later usage
cities.to_csv("dallas_cowboys_tweet_activity_2013.csv", index=False)
```

```python
# Adding population column to cities
population_data = {
        "city_state": [
        "Dallas, TX", "San Antonio, TX", "Houston, TX", "New York, NY", "Los Angeles, CA", "Austin, TX", "San Angelo, TX", "Fort Worth, TX", "Chicago, IL", "Philadelphia, PA",
        "Washington, DC", "Irving, TX", "Arlington, TX", "Atlanta, GA", "San Diego, CA", "Las Vegas, NV", "El Paso, TX", "New Orleans, LA", "Albuquerque, NM", "Denver, CO"
    ],
    'population': [
        1_300_000, 1_500_000, 2_400_000, 8_412_513,
        3_900_000, 1_000_000, 100_200, 1_000_000,
        2_700_000, 1_600_000, 710_100, 258_300, 406_000,
        525_100, 1_400_000, 686_700, 682_300, 357_500, 559_200,
        731_900

    ]
}
population_df = pd.DataFrame(population_data)
cites_w_pop = cities.head(20).copy()
cites_w_pop = cites_w_pop.merge(population_df, on='city_state', how='left')
cites_w_pop
```

```python
# add per capita count column
cites_w_pop['per_capita_tweets'] = (cites_w_pop['count'] / cites_w_pop['population'])
cites_w_pop = cites_w_pop.sort_values(by='per_capita_tweets', ascending=False).reset_index(drop=True)
cites_w_pop
```

```python
# graph per capita
fig_pc = px.scatter_geo(cites_w_pop,
                     lat="lat",
                     lon="lon",
                     size="per_capita_tweets",
                     hover_name="city_state",
                     hover_data={"count": True, "population": True, "per_capita_tweets": True, "lat": False, "lon": False},
                     title="Dallas Cowboys Tweet Activity by City Per Capita (2013)",
                     scope="usa")
fig_pc.update_geos(
    scope="usa"
)
fig_pc.update_layout(
    title_x=0.5,
    title_font_size=20
)
fig_pc.show()
```

```python
# save the per capita plot
fig_pc.write_html("dallas_cowboys_tweet_activity_per_capita_2013.html")
```
